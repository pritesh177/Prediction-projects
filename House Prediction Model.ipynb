# We want to make sure that CHAS is included as a feature in both sets

```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index,test_index in split.split(df,df['CHAS']):
    train_dataset=df.loc[train_index]
    test_dataset=df.loc[test_index]
    
```

# Evaluate model

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
house_predictions=linear.predict(df2)
lin_mse=mean_squared_error(df_label,house_predictions)
lin_rmse=np.sqrt(lin_mse)
```

# It seems that model has overfitted after use of Linear model and then decision tree to solve this  we use cross validation here

```python
from sklearn.model_selection import cross_val_score
```

```python
scores=cross_val_score(linear,df2,df_label,scoring='neg_mean_squared_error',cv=10)
rmse=np.sqrt(-scores)
```

```python
rmse
```

```python
def scores_ev(scores):
    print('scores: ',scores)
    print('mean: ',scores.mean())
    print('std: ',scores.std())
```

```python
scores_ev(rmse)
```
